그런데, 위의 것들은 자료구조와 알고리즘 의 측면에서만 얘기한 거구요, 알고리즘과 코딩을 익혀서 그 이후에 어디에 응용할지에 따라서 배우면 좋을 내용은 좀 달라집니다. 예를 들어 웹 프로그래밍 분야를 택하는 것과 시스템 프로그래밍을 택하는 것, 데이터 분석에 응용하려는 경우, 인공지능 솔루션을 만들려 하는 경우, 모바일 프로그래밍을 익히려는 경우, 등등, 프로그래밍 기초와 자료구조, 알고리즘 기초를 익힌 후 선택할 수 있는 길들은 아주 많이 있고 이 중 어느 길을 선택하느냐에 따라서 갖추면 좋을 역량의 종류 또한 달라집니다.

하지만 결론을 이야기하자면, 프로그래밍을 직업으로 삼으려는 경우라면 (단지 취미로 하려는 것이 아니라면) 자료구조의 내용 중 조금 심화된 것들 (예를 들어 위에 언급한 덩어리들 중 먼저의 것) 은 이후에 어떤 분야를 택하려는 경우에든 배워 두면 도움이 됩니다.
